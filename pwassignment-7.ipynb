```python
# 12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an
# example using the `abc` module.

"""ey points about ABCs:

    Defining Abstract Methods: ABCs define one or more abstract methods using 
    the @abstractmethod decorator. These methods must be implemented by the subclasses.

    Enforcing Method Implementation: Subclasses of ABCs must implement all abstract methods defined
    by the ABC. If a subclass fails to implement any abstract method, it will raise a TypeError when 
    an instance of the subclass is created.

    Relationship to Inheritance: ABCs are typically used as parent classes in inheritance hierarchies. 
    Subclasses inherit the abstract methods and are required to provide concrete implementations for them.

    Runtime Type Checking: ABCs can also be used for runtime type checking using the isinstance() function, 
    allowing you to check if an object belongs to a particular abstract class or its subclass."""

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    
    def area(self):
        return self.length * self.breadth

circle = Circle(5)
rectangle = Rectangle(4, 6)


print("Area of the circle:", circle.area())  
print("Area of the rectangle:", rectangle.area()) 
```

```python
# 2. Describe the difference between compile-time polymorphism and runtime polymorphism in Python.

"""    Compile-Time Polymorphism (Also known as Static Polymorphism or Early Binding):
        In statically-typed languages, compile-time polymorphism typically involves method overloading,
        where multiple methods with the same name but different parameter lists are defined within a class.
        The compiler determines which method to call based on the number and types of arguments provided at compile time.
        In Python, method overloading is not directly supported as it is in languages like C++ or Java. However, 
        you can achieve a form of method overloading using default parameter values or variable-length 
        argument lists (*args and **kwargs).
        Since Python is dynamically-typed and uses late binding, method resolution typically occurs 
        at runtime, making the concept of compile-time polymorphism less relevant in Python.

    Runtime Polymorphism (Also known as Dynamic Polymorphism or Late Binding):
        Runtime polymorphism occurs when the method to be executed is determined at runtime based on the actual
        type of the object being referenced.
        This is typically achieved through method overriding, where a subclass provides a specific implementation of a method 
        that is already defined in its superclass.
        In Python, method overriding is a common form of polymorphism. When a method is invoked on an object, Python looks
        for the method in the object's class. If the method is not found, Python looks for it in the superclass hierarchy
        until it finds a matching method to execute.
        Method overriding allows subclasses to provide specialized behavior while still maintaining a common interface
        with the superclass.
        Since method resolution occurs dynamically at runtime, Python supports runtime polymorphism, enabling flexibility
        and code reuse in object-oriented programming."""
```

```python
# 15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.
"""Data: The type of data being manipulated, which could be primitive types (integers, floats, etc.) 
or compound types (lists, dictionaries, etc.).

Operations: The set of operations that can be performed on the data. These operations define how the data 
can be accessed, modified, or manipulated. Examples of operations include insertion, deletion, searching, sorting, etc.

Properties: The properties or constraints that the data and operations must satisfy. These properties ensure 
the correctness and consistency of the data structure and its operations.


Separation of Concerns: ADTs allow programmers to separate the interface of a data structure from its implementation.
This separation enables modular design, where changes to the implementation do not affect the code that uses the data
structure.

Code Reusability: By defining clear interfaces through ADTs, programmers can reuse data structures in different parts
of their code or in different projects. This promotes code reusability and reduces code duplication.

Ease of Understanding: ADTs provide a high-level view of data structures, focusing on what operations can be performed 
and what properties hold true, rather than the specific details of how those operations are implemented. This abstraction
makes it easier for programmers to understand and reason about complex data structures.

Encapsulation: ADTs encapsulate the internal details of data structures, hiding them from the rest of the program.
This encapsulation helps in managing complexity and preventing unintended manipulation of data.

Standardization: ADTs provide standardized interfaces for common data structures, making it easier for programmers
to communicate and collaborate on software projects. Standard ADTs, such as stacks, queues, trees, and graphs, are widely
used across various domains."""

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from an empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise IndexError("peek from an empty stack")

    def size(self):
        return len(self.items)
```

```python
# 16. Create a Python class for a computer system, demonstrating abstraction by defining common methods
# (e.g., `power_on()`, `shutdown()`) in an abstract base class.

from abc import ABC, abstractmethod

class ComputerSystem(ABC):
    def __init__(self, name):
        self.name = name
        self.powered_on = False
    
    @abstractmethod
    def power_on(self):
        pass
    
    @abstractmethod
    def shutdown(self):
        pass

class DesktopComputer(ComputerSystem):
    def __init__(self, name):
        super().__init__(name)
    
    def power_on(self):
        if not self.powered_on:
            print(f"Powering on {self.name}...")
            self.powered_on = True
        else:
            print(f"{self.name} is already powered on.")
    
    def shutdown(self):
        if self.powered_on:
            print(f"Shutting down {self.name}...")
            self.powered_on = False
        else:
            print(f"{self.name} is already powered off.")

class LaptopComputer(ComputerSystem):
    def __init__(self, name):
        super().__init__(name)
    
    def power_on(self):
        if not self.powered_on:
            print(f"Booting up {self.name}...")
            self.powered_on = True
        else:
            print(f"{self.name} is already booted.")
    
    def shutdown(self):
        if self.powered_on:
            print(f"Shutting down {self.name}...")
            self.powered_on = False
        else:
            print(f"{self.name} is already powered off.")

if __name__ == "__main__":
    desktop = DesktopComputer("My Desktop")
    desktop.power_on()
    desktop.shutdown()
    desktop.shutdown()

    laptop = LaptopComputer("My Laptop")
    laptop.power_on()
    laptop.power_on()
    laptop.shutdown()
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods
# (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self):
        self.books = {}

    @abstractmethod
    def add_book(self, book_title, author):
        pass

    @abstractmethod
    def borrow_book(self, book_title):
        pass

    @abstractmethod
    def return_book(self, book_title):
        pass

class MyLibrary(LibrarySystem):
    def add_book(self, book_title, author):
        if book_title in self.books:
            print(f"{book_title} is already in the library.")
        else:
            self.books[book_title] = author
            print(f"{book_title} by {author} has been added to the library.")

    def borrow_book(self, book_title):
        if book_title in self.books:
            del self.books[book_title]
            print(f"You have borrowed {book_title}. Please return it within due date.")
        else:
            print(f"Sorry, {book_title} is not available in the library.")

    def return_book(self, book_title):
        if book_title in self.books:
            print(f"Thank you for returning {book_title}.")
        else:
            self.books[book_title] = None
            print(f"Invalid return: {book_title} was not borrowed from this library.")


library = MyLibrary()
library.add_book("The Great Gatsby", "F. Scott Fitzgerald")
library.add_book("To Kill a Mockingbird", "Harper Lee")

library.borrow_book("The Great Gatsby")
library.borrow_book("Moby Dick")

library.return_book("The Great Gatsby")
library.return_book("Moby Dick")

```

```python
# 11. How can you encapsulate and hide the details of composed objects in Python classes to maintain
# abstraction?

"""To encapsulate and hide the details of composed objects in Python classes, you can follow these principles:

Private Attributes: Use private attributes (attributes starting with a single underscore _) to indicate that they
should not be accessed directly from outside the class. This discourages direct access and encourages the use of getter
and setter methods for interaction.

Getter and Setter Methods: Provide getter and setter methods to access and modify the attributes of composed objects.
This allows you to control how the attributes are accessed and modified, enabling validation and abstraction.

Limited Interface: Expose only the necessary methods of composed objects to the outside world, hiding the internal
implementation details. This reduces complexity and provides a cleaner interface for interacting with the objects.

Abstraction: Define abstract base classes if appropriate, specifying the methods that must be implemented by subclasses.
This promotes code clarity and ensures that composed objects adhere to a common interface."""

class Engine:
    def __init__(self, horsepower):
        self._horsepower = horsepower

    def start(self):
        print("Engine started")

    def stop(self):
        print("Engine stopped")

    def get_horsepower(self):
        return self._horsepower


class Wheel:
    def __init__(self, size):
        self._size = size

    def rotate(self):
        print("Wheel rotating")

    def get_size(self):
        return self._size


class Transmission:
    def __init__(self, transmission_type):
        self._transmission_type = transmission_type

    def shift_gear(self, gear):
        print(f"Shifted to gear {gear}")

    def get_transmission_type(self):
        return self._transmission_type


class Car:
    def __init__(self, make, model, year, engine, wheels, transmission):
        self._make = make
        self._model = model
        self._year = year
        self._engine = engine
        self._wheels = wheels
        self._transmission = transmission

    def start(self):
        print(f"Starting the {self._make} {self._model} {self._year}...")
        self._engine.start()

    def stop(self):
        print(f"Stopping the {self._make} {self._model} {self._year}...")
        self._engine.stop()

    def drive(self):
        print(f"Driving the {self._make} {self._model} {self._year}...")
        self._transmission.shift_gear('Drive')
        for wheel in self._wheels:
            wheel.rotate()


    def get_engine_horsepower(self):
        return self._engine.get_horsepower()

    def get_wheel_sizes(self):
        return [wheel.get_size() for wheel in self._wheels]

    def get_transmission_type(self):
        return self._transmission.get_transmission_type()



if __name__ == "__main__":
    engine = Engine(horsepower=200)
    wheels = [Wheel(size=18) for _ in range(4)]
    transmission = Transmission(transmission_type="Automatic")
    my_car = Car(make="Toyota", model="Corolla", year=2022, engine=engine, wheels=wheels, transmission=transmission)

    my_car.start()
    my_car.drive()
    my_car.stop()

    print("Engine Horsepower:", my_car.get_engine_horsepower())
    print("Wheel Sizes:", my_car.get_wheel_sizes())
    print("Transmission Type:", my_car.get_transmission_type())
```

```python
# 12. Create a Python class for a university course, using composition to represent students, instructors, and
# course materials.


class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def __str__(self):
        return f"Student: {self.name} (ID: {self.student_id})"

class Instructor:
    def __init__(self, name, instructor_id):
        self.name = name
        self.instructor_id = instructor_id

    def __str__(self):
        return f"Instructor: {self.name} (ID: {self.instructor_id})"

class CourseMaterial:
    def __init__(self, title, content):
        self.title = title
        self.content = content

    def __str__(self):
        return f"Course Material: {self.title}"

class UniversityCourse:
    def __init__(self, course_code, course_title, instructor, students, course_materials):
        self.course_code = course_code
        self.course_title = course_title
        self.instructor = instructor
        self.students = students
        self.course_materials = course_materials

    def display_info(self):
        print(f"Course: {self.course_code} - {self.course_title}")
        print("Instructor:", self.instructor)
        print("Students:")
        for student in self.students:
            print("-", student)
        print("Course Materials:")
        for material in self.course_materials:
            print("-", material)



student1 = Student("Alice", "S123456")
student2 = Student("Bob", "S789012")

instructor = Instructor("Dr. Smith", "I987654")

material1 = CourseMaterial("Lecture Slides", "Introduction to Python programming.")
material2 = CourseMaterial("Lab Exercises", "Hands-on practice exercises for Python programming.")

students = [student1, student2]
materials = [material1, material2]

python_course = UniversityCourse("CS101", "Introduction to Python Programming", instructor, students, materials)
python_course.display_info()


```

```python
# 13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for
# tight coupling between objects.


"""Increased Complexity: As you compose more objects together to form a complex system, the overall 
complexity of the codebase can increase. Each component may have its own set of methods and attributes, 
leading to a larger codebase that may be harder to understand and maintain.

Tight Coupling: When objects are composed together, there's a risk of tight coupling between them, where 
changes to one object may require corresponding changes to other objects. This can make the system less flexible
and harder to modify or extend. For example, if the interface of a composed object changes, all objects that depend
on it may need to be updated accordingly.

Dependency Management: Managing dependencies between composed objects can be challenging. If one object depends 
on another, it may be difficult to replace or modify that dependency without affecting other parts of the system.
This can lead to a situation where changes in one component ripple through the entire system, causing unexpected
side effects.

Testing Complexity: Testing composed objects can be more complex compared to standalone objects. When testing a 
composed object, you may need to consider the interactions between its components, leading to more comprehensive
test cases and potentially increased testing effort.

Object Creation Overhead: Creating and initializing composed objects may incur overhead, especially if there are 
multiple levels of composition or if the composed objects require complex initialization logic. This can impact 
performance and memory usage, especially in resource-constrained environments."""


```

```python
# 14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes,
# and ingredients.


class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

    def __str__(self):
        return f"{self.quantity} {self.name}"

class Dish:
    def __init__(self, name, ingredients, price):
        self.name = name
        self.ingredients = ingredients
        self.price = price

    def __str__(self):
        return f"{self.name}: {', '.join(str(ingredient) for ingredient in self.ingredients)} - ${self.price:.2f}"

class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def __str__(self):
        return f"{self.name} Menu:\n" + "\n".join(str(dish) for dish in self.dishes)

class Restaurant:
    def __init__(self, name, menus):
        self.name = name
        self.menus = menus

    def display_menu(self, menu_name):
        for menu in self.menus:
            if menu.name == menu_name:
                print(menu)
                return
        print(f"Menu '{menu_name}' not found.")


ingredient1 = Ingredient("Tomato", 2)
ingredient2 = Ingredient("Cheese", 150)
ingredient3 = Ingredient("Bread", 1)

dish1 = Dish("Margherita Pizza", [ingredient1, ingredient2], 12.99)
dish2 = Dish("Grilled Cheese Sandwich", [ingredient2, ingredient3], 8.99)

menu1 = Menu("Lunch", [dish1, dish2])

restaurant = Restaurant("The Great Restaurant", [menu1])

restaurant.display_menu("Lunch")
```

```python
# 15. Explain how composition enhances code maintainability and modularity in Python programs.

"""    Encapsulation:
        Composition promotes encapsulation by allowing objects to hide their internal implementation details.
        Each object is responsible for a specific aspect of functionality, and the details of how it accomplishes
        that functionality are hidden from other parts of the program.
        Encapsulation reduces dependencies between different parts of the codebase, making it easier to understand, 
        modify, and maintain individual components without affecting the rest of the system.

    Modularity:
        Composition facilitates modularity by breaking down complex systems into smaller, more manageable components.
        Each component encapsulates a specific set of functionality and can be developed, tested, and maintained 
        independently.
        Modularity promotes code reuse, as components can be easily reused in different contexts or combined to 
        create new functionality. This reduces duplication and improves maintainability by promoting a more cohesive and understandable codebase.

    Flexibility:
        Composition offers flexibility in system design by allowing components to be easily replaced or extended 
        without impacting the overall structure or behavior of the system. New features can be added by composing 
        existing components or adding new ones, promoting adaptability to changing requirements.
        Components can be configured and combined in various ways to create different system configurations, enabling
        customization and scalability.

    Reduced Coupling:
        Composition reduces coupling between different parts of the codebase compared to other design patterns 
        like inheritance. Objects are not tightly bound to the implementation details of the objects they interact 
        with, leading to loosely coupled and more maintainable code.
        Components communicate through well-defined interfaces, promoting a clear separation of concerns and reducing
        the likelihood of unintended side effects when modifying or extending the code.

    Ease of Testing:
        Composition facilitates unit testing by allowing individual components to be tested in isolation. Since each
        component encapsulates a specific set of functionality, it can be tested independently of other components,
        making it easier to identify and fix bugs.
        Mocking and dependency injection techniques can be used to simulate the behavior of dependent components 
        during testing, ensuring that each component behaves as expected in different scenarios."""
```

```python
# 16. Create a Python class for a computer game character, using composition to represent attributes like
# weapons, armor, and inventory.

class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def __str__(self):
        return f"Weapon: {self.name} (Damage: {self.damage})"

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def __str__(self):
        return f"Armor: {self.name} (Defense: {self.defense})"

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_items(self):
        print("Inventory:")
        for item in self.items:
            print("-", item)

class GameCharacter:
    def __init__(self, name, weapon, armor, inventory):
        self.name = name
        self.weapon = weapon
        self.armor = armor
        self.inventory = inventory

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def display_info(self):
        print(f"Character: {self.name}")
        print("Equipped Weapon:", self.weapon)
        print("Equipped Armor:", self.armor)
        self.inventory.display_items()



sword = Weapon("Sword", 10)
shield = Armor("Shield", 5)

inventory = Inventory()
inventory.add_item("Health Potion")
inventory.add_item("Mana Potion")

character = GameCharacter("Hero", sword, shield, inventory)
character.display_info()
```

```python
# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
"""Independence of Lifecycles: In aggregation, the lifecycle of the parts is not tied to the 
lifecycle of the whole. This means that the parts can exist on their own even if the whole is destroyed.
In contrast, in simple composition, the parts typically cease to exist when the whole is destroyed.

Weaker Relationship: Aggregation implies a weaker form of ownership compared to simple composition.
While the whole object may contain references to its parts, it does not have full control over their
creation, destruction, or lifecycle management.

Flexibility and Reusability: Aggregation promotes greater flexibility and reusability since the parts
can be shared among multiple wholes or used independently in different contexts. This makes the codebase
more modular and easier to maintain.

Multiplicity: Aggregation often involves a "one-to-many" relationship, where one whole object can be
composed of multiple parts, and each part can be shared among multiple wholes. Simple composition, on 
the other hand, typically involves a "one-to-one" relationship, where each whole object has a single part."""


class Professor:
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, name):
        self.name = name
        self.professors = []  

    def add_professor(self, professor):
        self.professors.append(professor)


professor1 = Professor("John Doe")
professor2 = Professor("Jane Smith")


computer_science_dept = Department("Computer Science")
computer_science_dept.add_professor(professor1)
computer_science_dept.add_professor(professor2)


```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Furniture: {self.name}"

class Appliance:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Appliance: {self.name}"

class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def __str__(self):
        room_info = f"Room: {self.name}\n"
        room_info += "Furniture:\n"
        for item in self.furniture:
            room_info += f"- {item}\n"
        room_info += "Appliances:\n"
        for item in self.appliances:
            room_info += f"- {item}\n"
        return room_info

class House:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def display_house_info(self):
        print("House Information:")
        for room in self.rooms:
            print(room)


# Example usage
living_room = Room("Living Room")
living_room.add_furniture(Furniture("Sofa"))
living_room.add_furniture(Furniture("Coffee Table"))
living_room.add_appliance(Appliance("TV"))

kitchen = Room("Kitchen")
kitchen.add_furniture(Furniture("Dining Table"))
kitchen.add_furniture(Furniture("Chairs"))
kitchen.add_appliance(Appliance("Oven"))
kitchen.add_appliance(Appliance("Refrigerator"))

my_house = House()
my_house.add_room(living_room)
my_house.add_room(kitchen)

my_house.display_house_info()
```

```python
# 19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified
# dynamically at runtime?

from abc import ABC, abstractmethod


class Component(ABC):
    @abstractmethod
    def operation(self):
        pass

class ConcreteComponentA(Component):
    def operation(self):
        print("Concrete Component A operation")

class ConcreteComponentB(Component):
    def operation(self):
        print("Concrete Component B operation")


class Client:
    def __init__(self, component):
        self.component = component

    def perform_operation(self):
        self.component.operation()

if __name__ == "__main__":
   
    component_a = ConcreteComponentA()
    client = Client(component_a)
    client.perform_operation() 

 
    component_b = ConcreteComponentB()
    client.component = component_b
    client.perform_operation()  
```

```python
# 20. Create a Python class for a social media application, using composition to represent users, posts, and
# comments.

class Comment:
    def __init__(self, user, content):
        self.user = user
        self.content = content

    def __str__(self):
        return f"Comment by {self.user}: {self.content}"

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []

    def add_comment(self, user, content):
        comment = Comment(user, content)
        self.comments.append(comment)

    def __str__(self):
        post_info = f"Post by {self.user}: {self.content}\n"
        post_info += "Comments:\n"
        for comment in self.comments:
            post_info += f"- {comment}\n"
        return post_info

class User:
    def __init__(self, username):
        self.username = username

    def create_post(self, content):
        return Post(self.username, content)

    def __str__(self):
        return f"User: {self.username}"


class SocialMediaApp:
    def __init__(self):
        self.users = []

    def add_user(self, username):
        user = User(username)
        self.users.append(user)

    def create_post(self, username, content):
        for user in self.users:
            if user.username == username:
                return user.create_post(content)
        print(f"User '{username}' not found.")

# Example usage
app = SocialMediaApp()
app.add_user("Alice")
app.add_user("Bob")

post1 = app.create_post("Alice", "Hello world!")
post1.add_comment("Bob", "Nice post!")

post2 = app.create_post("Bob", "Good morning!")
post2.add_comment("Alice", "Have a great day!")

print(post1)
print(post2)
```
